## Détermination du nombre de jets privés atterrissant à Toulouse ###

```python
import pandas as pd
```

### Chargement de la base de données de Toulouse

```python
#Changer à la base de données de Toulouse 
#data_toulouse = data = pd.read_csv(r'data_toulouse/new_data_toulouse.zip',compression='zip', engine='python', error_bad_lines=False)
data_toulouse = pd.read_csv('new_data_toulouse.csv')       #fichier contenant 4 semaines de 4 mois (une semaine par mois, les mois sont Septembre-Octobre-Novembre-Décembre)
data_toulouse
```

### Exemple : nombre d'avions EasyJet détectées 

```python
list_jet = data_toulouse.jet_prive
list_id = []
nb = 0
for i in range(len(list_jet)) :
    if (list_jet[i].__contains__('Easyjet') and (data_toulouse['Flight_ID'][i] not in list_id)) : 
        nb += 1
        list_id.append(data_toulouse['Flight_ID'][i])
        print('Flight ID : ' + data_toulouse['Flight_ID'][i])
print("Le nombre d'avions détectés est : " + str(nb))
```

### Nombre de Jets 

```python
list_jet = data_toulouse.jet_prive
list_id = []
nb_jets = 0
for i in range(len(list_jet)) :
    if (list_jet[i].__contains__('Jet2') or list_jet[i].__contains__('Private') and (data_toulouse['Flight_ID'][i] not in list_id)) : 
        nb_jets += 1
        list_id.append(data_toulouse['Flight_ID'][i])
        print('Flight ID : ' + data_toulouse['Flight_ID'][i])
print("Le nombre d'avions détectés est : " + str(nb_jets))
```
